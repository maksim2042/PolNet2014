```python
def add_or_inc_edge(g,f,t):
    """
    Adds an edge to the graph IF the edge does not exist already.
    If it does exist, increment the edge weight.
    Used for quick-and-dirty calculation of projected graphs from 2-mode networks.
    """
    if g.has_edge(f,t):
        g[f][t]['weight']+=1
    else:
        g.add_edge(f,t,weight=1)
```

```python
def process_tweets(statuses):

    hashtag_network = net.Graph()
    retweet_network = net.DiGraph()
    texts = []
    
    for status in statuses:
        text = status['text']
        screen_name = status['user']['screen_name']
        hashtags = [h['text'].lower() for h in status['entities']['hashtags']]
        mentions = [h['screen_name'] for h in status['entities']['user_mentions']]
        
        for pair in permutations(hashtags, 2):
            add_or_inc_edge(hashtag_network,pair[0],pair[1])
            
        for mention in mentions:
            add_or_inc_edge(retweet_network, screen_name, mention)
            
        texts.append(text)
    return hashtag_network, retweet_network, texts
```

```python
hh,rt, texts = process_tweets(statuses)
```

```python
def trim_edges(g, weight=1):
    """
    Remove edges with weights less then a threshold parameter ("weight")
    """
    g2=net.Graph()
    for f, to, edata in g.edges(data=True):
        if edata['weight'] > weight:
            g2.add_edge(f,to,edata)
    return g2
```

```python
hh2 = trim_edges(hh, weight = 2)
```

```python
def trim_degrees(g, degree=1):
    """
    Trim the graph by removing nodes with degree less then value of the degree parameter
    Returns a copy of the graph, so it's non-destructive.
    """
    g2=g.copy()
    d=net.degree(g2)
    for n in g2.nodes():
        if d[n]<=degree: g2.remove_node(n)
    return g2
```
